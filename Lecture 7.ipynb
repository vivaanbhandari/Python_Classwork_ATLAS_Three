# Report Card Alphabet Grades

```python
report_card_df
```

```python
# If Total Grade is greater than or equal to 140 make the Status Column Value as Pass 
report_card_df.loc[report_card_df["total_grade"]>=140, "Status"] = "Pass"

# If Total Grade is lower than 140 make the Status Column Value as Fail
report_card_df.loc[report_card_df["total_grade"]<140, "Status"] = "Fail"

# Print the Report Card DataFrame
report_card_df
```

```python
# Making an Alphabet Grade Column
report_card_df.loc[(report_card_df["average_grade"] >=90), "Character_Grade"] = "A"
report_card_df.loc[(report_card_df["average_grade"] >=70) & (report_card_df["average_grade"]<90), "Character_Grade"] = "B"
report_card_df.loc[(report_card_df["average_grade"] >=50) & (report_card_df["average_grade"]<70), "Character_Grade"] = "C"
report_card_df.loc[(report_card_df["average_grade"] >=35) & (report_card_df["average_grade"]<50), "Character_Grade"] = "D"
report_card_df.loc[(report_card_df["average_grade"]<35), "Character_Grade"] = "F"
```

```python
report_card_df
```

```python
report_card_df = report_card_df.rename(columns={"Character_Grade": "Grades","average_grade": "Average Grades","total_grade": "Total Grade","biology_grades": "Biology Grades","math_grades": "Math Grades", "chemistry_grades": "Chemistry Grades","physics_grades": "Physics Grades", "student_names": "Student Names"})
report_card_df
```

```python
report_card_df.to_csv("Report Card.csv")
```
